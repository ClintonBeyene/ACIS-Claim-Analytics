#### Data cleaning filling missing values 

```python
# Remove rows with missing values in critical columns
df.dropna(subset=['PolicyID', 'TransactionMonth', 'IsVATRegistered', 'Citizenship', 'LegalType', 'Title', 'Language', 'Country', 'Province', 'PostalCode', 'MainCrestaZone', 'SubCrestaZone', 'ItemType', 'SumInsured', 'TermFrequency', 'CalculatedPremiumPerTerm', 'ExcessSelected', 'CoverCategory', 'CoverType', 'CoverGroup', 'Section', 'Product', 'StatutoryClass', 'StatutoryRiskType'], inplace=True)
```

```python
df['Bank'] = df['Bank'].fillna(df['Bank'].mode()[0])
```

```python
# Fill missing values in the 'AccountType' column with the mode
df.fillna({'AccountType': df['AccountType'].mode()[0]}, inplace=True)
```

#### since all of the banks are from south africa and the entry of citizenship of AF and ZW are very low we take to fill null values of citizenship with it's mode that are ZA.

```python
# Create a DataFrame indicating where values are missing
missing_df = df.isnull()

# Calculate the correlation matrix for missing values
missing_corr = missing_df.corr()

# Display the correlation matrix
print(missing_corr)
```

#### Check if some similar number of missing values with rows have similar rows are missing

```python
# Define the columns with missing values
columns_with_missing_values = ['mmcode', 'VehicleType','make', 'Model', 'Cylinders', 'cubiccapacity', 'kilowatts', 'bodytype', 'NumberOfDoors', 'VehicleIntroDate']

# Create a list to store the indices of missing rows
missing_row_indices = df[df['mmcode'].isnull()].index

# Loop through each column
for column in columns_with_missing_values[1:]:
    # Get the indices of missing rows in the current column
    missing_rows = df[df[column].isnull()].index
    
    # Intersect the current list of missing row indices with the new list
    missing_row_indices = missing_row_indices.intersection(missing_rows)

# Check if all columns are missing similar rows
if len(missing_row_indices) == len(df[df['mmcode'].isnull()]):
    print("All columns are missing similar rows.")
else:
    print("Not all columns are missing similar rows.")
```

```python
# Remove rows where 'mmcode' is missing (since all other columns have missing values in the same rows)
df.dropna(subset=['mmcode'], inplace=True)
```

#### Drop the column if no value exist, fill categorical data with mode, fill numerical data with mean

```python
# Define the columns with missing values
columns_with_missing_values = ['MaritalStatus', 'Gender', 'CustomValueEstimate', 'CapitalOutstanding', 'NewVehicle', 'WrittenOff', 'Rebuilt', 'Converted', 'CrossBorder', 'NumberOfVehiclesInFleet']

# Fill in the missing values with the mode of each column
for column in columns_with_missing_values:
    if df[column].count() == 0:
        df.drop(column, axis=1, inplace=True)
    else:
        if df[column].dtype == 'object':
           df[column] = df[column].fillna(df[column].mode().iloc[0])
        else:
           df[column] = df[column].fillna(df[column].mean())
```

##### Feature Engineering
###### Create New Features: Derive new features that might be useful for analysis

```python
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
```

```python
df['SumInsured'] = df['SumInsured'].astype(float)
```

```python
df['Year'] = df['TransactionMonth'].dt.year
```

```python
df['Month'] = df['TransactionMonth'].dt.month
```

###### Categorical encoding: convert categorical variables into numerical values.

```python
df = pd.get_dummies(df, columns=['IsVATRegistered', 'LegalType', 'Title', 'Language', 'Bank', 
                                 'AccountType', 'MaritalStatus', 'Gender', 'Country', 'Province', 
                                 'CoverCategory', 'CoverType', 'CoverGroup', 'Section', 'Product', 
                                 'StatutoryClass', 'StatutoryRiskType'])
```

 ##### Data Normalization
 ###### Normalize Numerical Features: Scale numerical features to a standard range.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['SumInsured_scaled'] = scaler.fit_transform(df[['SumInsured']])
df['CalculatedPremiumPerTerm_scaled'] = scaler.fit_transform(df[['CalculatedPremiumPerTerm']])
df['TotalPremium_scaled'] = scaler.fit_transform(df[['TotalPremium']])
df['TotalClaims_scaled'] = scaler.fit_transform(df[['TotalClaims']])
```
